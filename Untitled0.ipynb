Here we set up the data into the learning (800) and test (200) sets, and also make the y values iterate from 0 and 1 for good and bad respectively for ease of calc of true values later

```python
def ohet(original,new) :
  enc = OneHotEncoder(handle_unknown='ignore',sparse_output = False).set_output(transform='pandas')

  trainhot=enc.fit_transform(original)
  testhot=enc.transform(new)
  return trainhot,testhot
```

Here we are using the OneHotEncoder in Scikit to be able to make categorical values into numerical indicator values to use in later steps.

```python
def fix(original,new):
  mu=np.mean(original)
  sig=np.std(original)
  return (original-mu)/sig, (new-mu)/sig
```

Here we normalize the characteristics values from the OneHotEncoder to make it easier later on

```python
FN = FP = TP = TN = 0
for i in range(len(y_true)):
    if y_true[i]==1 and y_pred[i]==1:
        TP += 1
    elif y_true[i]==0 and y_pred[i]==0:
        TN += 1
    elif y_true[i]==1 and y_pred[i]==0:
        FN += 1
    else:
        FP += 1

precision = TN / (TN + FN)
print("Baseline precision:", precision)

accuracy = (TP + TN) / (TP + TN + FP + FN)
print("Baseline accuracy:", accuracy)

alpha = (FP)/(FP+TN)
print("Alpha:",alpha)

beta=(FN)/(FN+TP)
print("Beta:",beta)
```

Here we have 1 values equal to positives and 0 equal to negatives. When the true values set and prediction values set of the trial values we can compare to find the number of true positives, true negatives, false positives, and false negatives.


```python
cost_matrix = {
    (0,0): 0,
    (0,1): 1,
    (1,0): 5,
    (1,1): 0
}

baseline_cost = np.sum([
    cost_matrix[(a, p)] for a, p in zip(y_true, y_pred)
])

print("Baseline cost with threshold = 0.5:",baseline_cost)
```

This is our baseline cost with the cost matrix of 1 for FP and 5 for FN with the .5 default threshold porbibility

```python
y_pred_proba = log_reg_model.predict_proba(Xtenew)[:,1]

thresholds = np.linspace(0,1,100)
costs_per_threshold = []

for threshold in thresholds:
    y_pred_thresholded = (y_pred_proba >= threshold).astype(int)
    current_cost = np.sum([cost_matrix[(a,p)] for a,p in zip(y_true, y_pred_thresholded)])
    costs_per_threshold.append(current_cost)

optimal_index = np.argmin(costs_per_threshold)
optimal_threshold = thresholds[optimal_index]
min_cost = costs_per_threshold[optimal_index]

print("Optimal threshold: ",optimal_threshold)
print("Minimum cost: ", min_cost)

y_pred_opt = (y_pred_proba >= optimal_threshold).astype(int)
optimal_cost = np.sum([cost_matrix[(a,p)] for a,p in zip(y_true, y_pred_opt)])
```

This gives us a optimal porblity for the cost matrix and gives the one with least cost as well



```python
FN = FP = TP = TN = 0
for i in range(len(y_true)):
    if y_true[i]==1 and y_pred_opt[i]==1:
        TP += 1
    elif y_true[i]==0 and y_pred_opt[i]==0:
        TN += 1
    elif y_true[i]==1 and y_pred_opt[i]==0:
        FN += 1
    else:
        FP += 1

precision_opt = TN / (TN + FN)
print("Optimized precision :", precision_opt)

accuracy_opt = (TP + TN) / (TP + TN + FP + FN)
print("Optimized accuracy:", accuracy_opt)

alpha = (FP)/(FP+TN)
print("Alpha:",alpha)

beta=(FN)/(FN+TP)
print("Beta:",beta)
```
